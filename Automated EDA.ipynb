```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
```

```python
import sqlalchemy
import seaborn as sns
```

```python
def handle_missing_values(data):
    
    data.fillna(data.median(), inplace=True)
    return data
```

```python
def create_visualizations(data):
    
    for col in data.columns:
        
        
        #not to include id columns
        if('id' not in col.lower()):
            
            
            
        
        
            if (data[col].dtype == 'int64' or data[col].dtype == 'float64'):
                plt.figure(figsize=(10, 5))

                #  Histogram plot vis
                plt.subplot(1, 3, 1)
                plt.hist(data[col])
                plt.title(f'Histogram of {col}')
                plt.xlabel(col)
                plt.ylabel('Frequency')

                #   Scatter Plot vis
                plt.subplot(1, 3, 2)
                plt.scatter(data.index, data[col])
                plt.title(f'Scatter Plot of {col}')
                plt.xlabel('Frequency')
                plt.ylabel(col)

                #Box Plot vis
                plt.subplot(1, 3, 3)
                data[col].plot(kind='box')
                plt.title(f'Box Plot of {col}')
                plt.ylabel(col)

                plt.tight_layout()
                plt.show()
                
            elif data[col].dtype =='object':
                #bar plot for category data
                
                data[col].value_counts().plot(kind='bar')
                plt.title(f'Bar Plot of {col}')
                plt.xlabel(col)
                plt.ylabel('Count')
                plt.xticks(rotation=45)
                plt.show()    


            else:
                print(f"Column '{col}' is not int or float,str, no visulization.")
```

```python
def load_data(data_source):
    if data_source.endswith('.csv'):
        return pd.read_csv(data_source)
    elif data_source.endswith('.xlsx'):
        return pd.read_excel(data_source)
    elif data_source.endswith('.sql'):
        engine = sqlalchemy.create_engine(data_source)
        return pd.read_sql_query('SELECT * FROM your_table', engine)
    else:
        raise ValueError("Unsupported data source")
```

```python
def EDA_program():
    
    file_path = input("please Enter the file path: ")
    loaded_data=load_data(file_path)
    cleaned_data = handle_missing_values(loaded_data)
    create_visualizations(cleaned_data)

    
    
    
```

```python
EDA_program()
```
